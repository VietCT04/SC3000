```python
import numpy as np
# Workaround for some NumPy versions where np.bool8 might not exist
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_

import gym
from gym import logger as gymlogger
from gym.wrappers import RecordVideo
gymlogger.set_level(40)  # Only show error messages

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import glob
import io
import base64
from IPython.display import HTML
from IPython import display as ipythondisplay

# ====================================================
# Define the Policy Network (used for all tasks)
# ====================================================
class PolicyNetwork(nn.Module):
    def __init__(self, state_size, action_size, hidden_size=128):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(state_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, action_size)
    
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return torch.softmax(x, dim=-1)

# ====================================================
# Environment Setup and Seed Initialization
# ====================================================
# For training and evaluation we use a non-rendering environment.
env = gym.make('CartPole-v1', render_mode=None)
state, _ = env.reset(seed=42)
env.action_space.seed(42)
env.observation_space.seed(42)
np.random.seed(42)
torch.manual_seed(42)

# Instantiate the policy network and optimizer
policy_net = PolicyNetwork(state_size=4, action_size=2)
optimizer = optim.Adam(policy_net.parameters(), lr=0.01)

# ====================================================
# Utility Function: Select Action (Stochastic Policy)
# ====================================================
def select_action(state):
    state_tensor = torch.FloatTensor(state)
    probs = policy_net(state_tensor)
    dist = torch.distributions.Categorical(probs)
    action = dist.sample()
    return action.item(), dist.log_prob(action)

# ====================================================
# Task 1: Demonstrate the Agent on a Random State
# ====================================================
print("----- Task 1: Demonstrate the Agent on a Random State -----")
print("Sampled state:", state)
action, log_prob = select_action(state)
print("Chosen action:", action)
# (Note: At this stage, the policy network is untrained.)

# ====================================================
# Task 2: Train the Agent Using REINFORCE & Evaluate It
# ====================================================
def train_policy(episodes=1000, gamma=0.99):
    policy_net.train()  # Set the network to training mode
    training_rewards = []  # To store cumulative reward per episode

    for ep in range(episodes):
        state, _ = env.reset(seed=42)  # Reset environment at the start of each episode
        log_probs = []  # Store log probabilities of the actions taken
        rewards = []    # Store rewards for each timestep
        total_reward = 0
        done = False
        
        # Run one episode
        while not done:
            action, log_prob = select_action(state)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            log_probs.append(log_prob)
            rewards.append(reward)
            state = next_state
            total_reward += reward
        
        # Compute discounted rewards for the episode
        discounted_rewards = []
        R = 0
        for r in reversed(rewards):
            R = r + gamma * R
            discounted_rewards.insert(0, R)
        discounted_rewards = torch.FloatTensor(discounted_rewards)
        # Normalize the discounted rewards to improve training stability
        discounted_rewards = (discounted_rewards - discounted_rewards.mean()) / (discounted_rewards.std() + 1e-9)
        
        # Compute the policy loss (REINFORCE loss)
        loss = 0
        for log_prob, R in zip(log_probs, discounted_rewards):
            loss -= log_prob * R
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        training_rewards.append(total_reward)
        if (ep + 1) % 100 == 0:
            print(f"Episode {ep+1}: Total Reward = {total_reward}")
    
    return training_rewards

print("\n----- Task 2: Training the Agent -----")
training_rewards = train_policy(episodes=1000)

# Plot training progress: Cumulative reward per episode
plt.figure(figsize=(10, 5))
plt.plot(training_rewards)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Training Progress: Total Reward per Episode')
plt.grid(True)
plt.show()

# Evaluate the trained agent over 100 episodes
def evaluate_policy(episodes=100):
    policy_net.eval()  # Set network to evaluation mode
    eval_rewards = []
    for ep in range(episodes):
        state, _ = env.reset(seed=42)
        total_reward = 0
        done = False
        while not done:
            state_tensor = torch.FloatTensor(state)
            probs = policy_net(state_tensor)
            # Use a greedy policy (choose action with highest probability)
            action = int(torch.argmax(probs).item())
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            state = next_state
            total_reward += reward
        eval_rewards.append(total_reward)
    avg_reward = np.mean(eval_rewards)
    return eval_rewards, avg_reward

eval_rewards, avg_reward = evaluate_policy(episodes=100)
print("\nEvaluation over 100 episodes:")
print("Average Reward:", avg_reward)

# ====================================================
# Task 3: Render One Episode and Record a Video
# ====================================================
# Utility function to display the recorded video (works in Jupyter/IPython)
def show_video():
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[0]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        ipythondisplay.display(HTML(data='''<video alt="video" autoplay loop controls style="height: 400px;">
            <source src="data:video/mp4;base64,{0}" type="video/mp4" />
         </video>'''.format(encoded.decode('ascii'))))
    else:
        print("Could not find video")

# For video recording, create an environment that produces renderable frames.
video_env = RecordVideo(gym.make('CartPole-v1', render_mode='rgb_array'),
                        video_folder='video',
                        episode_trigger=lambda episode_id: True)

state, _ = video_env.reset(seed=42)
done = False

# Run one episode using a greedy policy (selecting the action with the highest probability)
while not done:
    state_tensor = torch.FloatTensor(state)
    probs = policy_net(state_tensor)
    action = int(torch.argmax(probs).item())
    state, reward, terminated, truncated, _ = video_env.step(action)
    done = terminated or truncated

video_env.close()  # Finalize and save the video

print("\n----- Task 3: Displaying Recorded Episode Video -----")
show_video()
```

```python
# cartpole_dqn_improved.py

# Workaround for some NumPy versions that lack np.bool8
import numpy as np
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_

import gym
from gym import logger as gymlogger
from gym.wrappers import RecordVideo
gymlogger.set_level(40)  # Only show error messages

import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque

# For video display in Jupyter/IPython:
import glob, io, base64
from IPython.display import HTML
from IPython import display as ipythondisplay

# --------------------------------------------------
# Define the Q-Network (Deep Q-Network)
# --------------------------------------------------
class QNetwork(nn.Module):
    def __init__(self, state_size, action_size, hidden_size=64):
        super(QNetwork, self).__init__()
        self.fc1 = nn.Linear(state_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, action_size)
        
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)  # output raw Q-values for each action

# --------------------------------------------------
# Replay Buffer for Experience Replay
# --------------------------------------------------
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)
        
    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))
        
    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = map(np.array, zip(*batch))
        return states, actions, rewards, next_states, dones
        
    def __len__(self):
        return len(self.buffer)

# --------------------------------------------------
# Epsilon-Greedy Action Selection for DQN
# --------------------------------------------------
def select_action(state, q_network, epsilon, action_size):
    if random.random() < epsilon:
        return random.randrange(action_size)
    else:
        state_tensor = torch.FloatTensor(state).unsqueeze(0)
        with torch.no_grad():
            q_values = q_network(state_tensor)
        return int(torch.argmax(q_values).item())

# --------------------------------------------------
# Environment Setup and Initialization
# --------------------------------------------------
env = gym.make('CartPole-v1', render_mode=None)
state, _ = env.reset(seed=42)  # Initial state for demonstration
env.action_space.seed(42)
env.observation_space.seed(42)
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)

state_size = env.observation_space.shape[0]  # 4 for CartPole
action_size = env.action_space.n             # 2 possible actions

# Instantiate Q-network and target network
q_network = QNetwork(state_size, action_size)
target_network = QNetwork(state_size, action_size)
target_network.load_state_dict(q_network.state_dict())

# Optimizer & Replay Buffer
optimizer = optim.Adam(q_network.parameters(), lr=1e-4)
replay_buffer = ReplayBuffer(capacity=10000)

# --------------------------------------------------
# Task 1: Q-Network Demonstration on a Random State
# --------------------------------------------------
print("----- Task 1: Q-Network demonstration on a random state -----")
print("Sampled state:", state)
with torch.no_grad():
    q_vals = q_network(torch.FloatTensor(state).unsqueeze(0))
print("Q-values:", q_vals)
chosen_action = int(torch.argmax(q_vals).item())
print("Chosen action:", chosen_action)

# --------------------------------------------------
# Task 2: Train the DQN Agent for 1000 Episodes
# --------------------------------------------------
def train_dqn(env, q_network, target_network, optimizer, replay_buffer,
              num_episodes=1000, batch_size=128, gamma=0.99,
              epsilon_start=1.0, epsilon_final=0.01, epsilon_decay=2000,
              target_update_freq=500):
    """
    Trains the DQN agent for num_episodes using a Double DQN update.
    Returns a list of total rewards per episode.
    """
    episode_rewards = []
    steps_done = 0
    epsilon = epsilon_start
    
    for episode in range(num_episodes):
        state, _ = env.reset()  # Varied initial conditions
        total_reward = 0
        done = False
        
        while not done:
            # Decay epsilon (slower decay for longer exploration)
            epsilon = epsilon_final + (epsilon_start - epsilon_final) * np.exp(-1.0 * steps_done / epsilon_decay)
            action = select_action(state, q_network, epsilon, action_size)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            
            # Store transition in replay buffer
            replay_buffer.push(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            steps_done += 1
            
            # Update Q-network if enough samples are available
            if len(replay_buffer) >= batch_size:
                states, actions, rewards, next_states, dones = replay_buffer.sample(batch_size)
                
                states = torch.FloatTensor(states)
                actions = torch.LongTensor(actions)
                rewards = torch.FloatTensor(rewards)
                next_states = torch.FloatTensor(next_states)
                dones = torch.FloatTensor(dones.astype(np.float32))
                
                # Current Q-values
                current_q = q_network(states).gather(1, actions.unsqueeze(1)).squeeze(1)
                
                # Double DQN update:
                # 1) Choose best next actions using the online network.
                with torch.no_grad():
                    best_next_actions = q_network(next_states).argmax(dim=1, keepdim=True)
                # 2) Evaluate those actions using the target network.
                with torch.no_grad():
                    next_q = target_network(next_states).gather(1, best_next_actions).squeeze(1)
                target_q = rewards + gamma * next_q * (1 - dones)
                
                loss = nn.MSELoss()(current_q, target_q)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
            
            # Update the target network every target_update_freq steps
            if steps_done % target_update_freq == 0:
                target_network.load_state_dict(q_network.state_dict())
        
        episode_rewards.append(total_reward)
        if (episode + 1) % 50 == 0:
            print(f"Episode {episode+1}: Total Reward = {total_reward}")
    
    return episode_rewards

print("\n----- Task 2: Training the DQN Agent for 1000 Episodes -----")
training_rewards = train_dqn(env, q_network, target_network, optimizer, replay_buffer, 
                             num_episodes=1000)

# We don't care about the training graph, so we won't optimize for it.

# Evaluate the trained agent over 100 episodes (greedy policy)
def evaluate_dqn(env, q_network, episodes=100):
    """
    Runs the agent for 'episodes' times in a greedy fashion (argmax) and returns the average total reward.
    """
    q_network.eval()
    eval_rewards = []
    for _ in range(episodes):
        state, _ = env.reset()
        total_reward = 0
        done = False
        while not done:
            with torch.no_grad():
                q_vals = q_network(torch.FloatTensor(state).unsqueeze(0))
            action = int(torch.argmax(q_vals).item())
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            state = next_state
            total_reward += reward
        eval_rewards.append(total_reward)
    avg_reward = np.mean(eval_rewards)
    return eval_rewards, avg_reward

eval_rewards, avg_reward = evaluate_dqn(env, q_network, episodes=100)
print("\nEvaluation over 100 episodes:")
print("Average Reward:", avg_reward)

# --------------------------------------------------
# Task 3: Record and Display a Video of One Episode
# --------------------------------------------------
def show_video():
    """Displays the first MP4 file in the 'video' folder (works in Jupyter/IPython)."""
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[0]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        ipythondisplay.display(HTML(data='''<video width="400" controls>
                    <source src="data:video/mp4;base64,{0}" type="video/mp4">
                 </video>'''.format(encoded.decode('ascii'))))
    else:
        print("Could not find video")

video_env = RecordVideo(gym.make('CartPole-v1', render_mode='rgb_array'),
                        video_folder='video',
                        episode_trigger=lambda episode_id: True)

state, _ = video_env.reset()
done = False
while not done:
    with torch.no_grad():
        q_vals = q_network(torch.FloatTensor(state).unsqueeze(0))
    action = int(torch.argmax(q_vals).item())
    state, reward, terminated, truncated, _ = video_env.step(action)
    done = terminated or truncated

video_env.close()
print("\n----- Task 3: Displaying Recorded Episode Video -----")
show_video()
```
